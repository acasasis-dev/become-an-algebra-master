## Exercise 6: Domain and Range

> 1. Find the domain of $f(x)$.

\begin{gather*}
    f(x) = \frac{3}{x(x + 1)} + x^2 \\ \\

    x \ne 0, -1 \\
    f(x) \ne 0
\end{gather*}

> 2. Find the domain and range of the point set.

\begin{gather*}
    (-1, -3), \ \ (0, 5), \ \ (-3, 6), \ \ (0, -3) \\ \\

    \text{domain: } 0, -1, -3 \\
    \text{range: } -3, 5, 6
\end{gather*}

> 3. Find the domain and range of $g(x)$.

\begin{gather*}
    g(x) = \frac{\sqrt{x - 2}}{3} \\ \\

    x \ge 2 \\
    g(x) \ge 0
\end{gather*}

> 4. Find the domain and range of the function.

\begin{gather*}
    f(x) = \frac{2}{x} + 1 \\ \\
\end{gather*}

> 5. Find the domain and range of $g(x)$.

\begin{gather*}
    g(x) = -x^2 + 5 \\ \\
\end{gather*}

> 6. What is the domain and range of the graph? Hint: An empty circle
indicates that exact point *is not* included as part of the graph, while a solid
circle indicates that exact point *is* included as part of the graph.
