> 4. Show that the function is one-to-one by showing that $f(a) = f(b)$ leads
to $a = b$.

\begin{gather*}
    f(x) = 3x − 4 \\ \\

    3a - 4 = 3b - 4 \\
    3a = 3b \\
    a = b
\end{gather*}

> 5. Show that the function is one-to-one by showing that $f(a) = f(b)$ leads
to $a = b$.

\begin{gather*}
    f(x) = \frac{x + 1}{x - 5} \\ \\

    \frac{a + 1}{a - 5} = \frac{b + 1}{b - 5} \\
    (a + 1)(b - 5) = (a - 5)(b + 1) \\
    ab - 5a + b - 5 = ab + a - 5b - 5 \\
    ab - 5a + b = ab + a - 5b \\
    -5a + b = a - 5b \\
    a = b
\end{gather*}
