## Exercise 1: Parallel and Perpendicular Lines

```python
import matplotlib.pyplot as plt
import numpy as np

def plot(*args):
    x = np.arange(-20, 20)
    for f in args:
        y = f(x)
        plt.plot(x, y)

    plt.axhline(0, color="black", linewidth=2)
    plt.axvline(0, color="black", linewidth=2)
    plt.xlim(-20, 20)
    plt.ylim(-20, 20)
    plt.grid()
    plt.show()
```

> 1. Determine if the lines are parallel, perpendicular, or neither.

\begin{gather*}
    y = 2x + 5 \\
    2y + x = -3 \\ \\

    2y = -x - 3 \\
    \frac{2y = -x - 3}{2} \\
    y = -\frac{1}{2}x - \frac{3}{2}
\end{gather*}

```python
def f_1(x):
    return 2*x + 5

def f_2(x):
    return -0.5*x - 1.5

plot(f_1, f_2)
```

The $2$ lines are **PERPENDICULAR**

> 2. Determine if the lines are parallel, perpendicular, or neither.

    \frac{10y = 15x + 20}{5} \\
    2y = 3x + 4 \\
    \frac{2y = 3x + 4}{2} \\
    y = \frac{3}{2}x + 2
\end{gather*}

```python
def f_1(x):
	return 5*x + 1

def f_2(x):
	return (3/2)*x + 2

plot(f_1, f_2)
```

> 3. Write the equation of the line with a $y$ -intercept of $3$ thats parallel to
$7x + 3y = 12$.

\begin{gather*}
    7x + 3y = 12 \\
    3y = -7x + 12 \\
    \frac{3y = -7x + 12}{3} \\
    y = -\frac{7}{3}x + 4 \\
\end{gather*}

```python
def f_1(x):
	return -(7/3)*x + 4

def f_2(x):
	return -(7/3)*x - 3

plot(f_1, f_2)
```

The answer is $y = -\frac{7}{3}x - 3$

> 4. Write the equation of the line passing through $(3,6)$ and perpendicular
to $6x + 2y = 4$.

\begin{gather*}
    6x + 2y = 4 \\
    2y = -6x + 4 \\
    y = -3x + 2 \\ \\

    6 = \frac{1}{3}(3) + b \\
    6 = 1 + b \\
    b = 6 - 1 \\
    b = 5 \\ \\

    y = \frac{1}{3}x + 5
\end{gather*}

```python
def f_1(x):
	return -3*x + 2

def f_2(x):
	return (1/3)*x + 5

plot(f_1, f_2)
```
