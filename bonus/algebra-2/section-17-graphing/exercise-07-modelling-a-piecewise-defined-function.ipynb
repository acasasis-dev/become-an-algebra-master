## Exercise 07: Modelling A Piecewise-Defined Function

```python
import matplotlib.pyplot as plt
import numpy as np

def plot(*args, ranges=None):
    if ranges:
        for i, r in enumerate(ranges):
            start, end = r
            x = np.arange(start, end)
            y = args[i](x)
            plt.plot(x, y)
    else:
        x = np.arange(-20, 20)
        for f in args:
            y = f(x)
            plt.plot(x, y)

    plt.axhline(0, color="black", linewidth=2)
    plt.axvline(0, color="black", linewidth=2)
    plt.xticks(np.arange(-20, 20, 1))
    plt.yticks(np.arange(-20, 20, 1))
    plt.xlim(-20, 20)
    plt.ylim(-20, 20)
    plt.grid()
    plt.show()  
```

> 1. Find the equation of the piecewise function.



\begin{gather*}
    y = (x + 3)^2 \\
    y = x^2 + 6x + 9 \\ \\

    f(x) = \begin{cases}
        x^2 + 6x + 9 & x < 0 \\
        x + 9 & x \ge 0
    \end{cases}
\end{gather*}


> 2. Find the equation of the piecewise function.



\begin{gather*}
    f(x) = \begin{cases}
        -2 & x < -3 \\
        \frac{2}{3}x & -3 \le x \le 6 \\
        4 & x > 6
    \end{cases}
\end{gather*}

> 3. Find the equation of the piecewise function.



\begin{gather*}
    f(x) = \begin{cases}
        -x^2 & x < 1 \\
        1 & x = 1 \\
        x + 2 & x > 1
    \end{cases}
\end{gather*}

> 4. Graph the piecewise function.

\begin{gather*}
    f(x) = 
    \begin{cases}
        x^2 & x < -1 \\
        \frac{1}{2}x - 3 & x \ge -1
    \end{cases}
\end{gather*}

```python
def f_1(x):
	return x**2

def f_2(x):
	return (1/2)*x - 3

plot(f_1, f_2, ranges=[(-20, 0), (-1, 20)])
```

> 5. Graph the piecewise function.

\begin{gather*}
    f(x) = 
    \begin{cases}
        x^2 - 2 & x < 0 \\
        4 & x = 0 \\
        -x^2 + 8 & x > 0
    \end{cases}
\end{gather*}

```python
def f_1(x):
    return x**2 - 2

def f_2(x):
    return 4

def f_3(x):
    return -x**2 + 8

plot(f_1, f_2, f_3, ranges=[(-20, 1), (0, 1), (0, 20)])
```

> 6. Graph the piecewise function.

\begin{gather*}
    f(x) =
    \begin{cases}
        -x + 1 & x < -4 \\
        5 & -4 < x \le 3 \\
        -2x + 11 & x > 3
    \end{cases}
\end{gather*}

```python
def f_1(x):
	return -x + 1

def f_2(x):
	return [5] * 8

def f_3(x):
	return -2*x + 11

plot(f_1, f_2, f_3, ranges=[(-20, -3), (-4, 4), (3, 20)])
```
