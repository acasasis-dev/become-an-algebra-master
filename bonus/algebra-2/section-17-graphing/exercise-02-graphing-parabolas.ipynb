## Exercise 2: Graphing Parabolas

```python
import matplotlib.pyplot as plt
import numpy as np

def plot(*args):
    x = np.arange(-20, 20)
    for f in args:
        y = f(x)
        plt.plot(x, y)

    plt.axhline(0, color="black", linewidth=2)
    plt.axvline(0, color="black", linewidth=2)
    plt.xticks(np.arange(-10, 10, 1))
    plt.yticks(np.arange(-10, 10, 1))
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.grid()
    plt.show()
```

> 1. Write the equation in vertex form.

\begin{gather*}
    y = −2x^2 + 24x − 68 \\ \\

    y = -2(x^2 - 12x) - 68 \\ \\
    \left(-\frac{12}{2}\right)^2 = -6^2 = 36 \\
    y - 2(36) = -2(x^2 + 12x + 36) - 68 \\
    y - 72 = -2(x + 6)^2 - 68 \\
    y = -2(x + 6)^2 - 68 + 72 \\
    y = -2(x + 6)^2 + 4
\end{gather*}

> 2. Find the vertex and axis of symmetry of $y = x^2 + 5x + 6$.

\begin{gather*}
    y = x^2 + 5x + 6 \\ \\

    \left(\frac{5}{2}\right)^2 = \frac{25}{4} \\
    y + \frac{25}{4} = (x^2 + 5x + \frac{25}{4}) + 6 \\
    y = (x^2 + 5x + \frac{25}{4}) + 6 - \frac{25}{4} \\
    y = (x + \frac{5}{2})^2 + \frac{24}{4} - \frac{25}{4} \\
    y = (x + \frac{5}{2})^2 - \frac{1}{4}
\end{gather*}

The vertex is $(-\frac{5}{2}, -\frac{1}{4})$ and the axis of symmetry is $-\frac{5}{2}$

> 3. Find the vertex and axis of symmetry of $y = 3(x + 2)^2 + 6$.

> 4. Using the graph, find the equation of the parabola in standard form.



\begin{gather*}
    y = a(x - h)^2 + k \\
    y = a(x - 1)^2 + -2 \\
    y = a(x - 1)^2 - 2 \\ \\

    x = 0, y = 0 \\
    0 = a(0 - 1)^2 - 2 \\
    0 = a(-1)^2 - 2 \\
    0 = a - 2 \\
    a = 2 \\ \\

    y = 2(x - 1)^2 - 2 \\
    y = 2(x - 1)(x - 1) - 2 \\
    y = 2(x^2 - 2x + 1) - 2 \\
    y = 2x^2 - 4x + 2 - 2 \\
    \boxed{y = 2x^2 - 4x}
\end{gather*}

> 5. Complete the square to graph the parabola $y = x^2 + 6x + 5$.

\begin{gather*}
    y = x^2 + 6x + 5 \\ \\

    \left(\frac{6}{2}\right)^2 = 3^2 = 9 \\
    y + 9 = (x^2 + 6x + 9) + 5 \\
    y = (x + 3)^2 + 5 - 9 \\
    \boxed{y = (x + 3)^2 - 4} \\ \\

    x^2 + 6x + 5 = 0 \\
    (x + 5)(x + 1) = 0 \\ \\

    (-5, 0), (-3, -4), (-1, 0)
\end{gather*}

> 6. Complete the square to graph $y = − x^2 − 4x − 6$.

\begin{gather*}
    y = − x^2 − 4x − 6 \\
    y = -(x^2 + 4x) - 6 \\ \\

    \left(\frac{4}{2}\right)^2 = 2^2 = 4 \\
    y = -(x^2 + 4x + 4) - 6 + 4 \\
    \boxed{y = -(x + 2)^2 - 2} \\ \\

    x = 0 \\
    y = -0^2 - 4(0) - 6 \\
    y = -6 \\ \\

    (-2, -2), (0, -6)
\end{gather*}
