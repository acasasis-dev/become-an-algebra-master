## Exercise 3: Graphing Inequalities On A Number Line

> 1. Give two inequalities that, when graphed on a number line, have open
circles at $x = 3$.

\begin{gather*}
    x < 3
\end{gather*}



\begin{gather*}
    x > 3
\end{gather*}



> 2. Graph the inequality on a number line.

\begin{gather*}
    −2x < 4 \\ \\

    \frac{-2x < 4}{-2} \\
    x > -2
\end{gather*}



> 3. Graph the inequality on a number line.

\begin{gather*}
    x − 1 \ge 3 \\ \\

    x \ge 3 + 1 \\
    x \ge 4
\end{gather*}



> 4. Graph the inequality on a number line.

\begin{gather*}
    5(−x + 3) < − 3x + 7 \\ \\

    -5x + 15 < -3x + 7 \\
    -5x + 3x < 7 - 15 \\
    -2x < -8 \\ \\

    \frac{-2x < -8}{-2} \\
    x > 4
\end{gather*}



> 5. What’s wrong with this graph of $x > 1$?



The line is correct but the circle at point $1$ is not, it should be an empty circle

> 6. Graph the inequality on a number line.

\begin{gather*}
    5(x + 7) − x \ge 3(x + 10) + 6 \\ \\

    5x + 35 - x \ge 3x + 30 + 6 \\
    5x - x - 3x \ge 30 + 6 - 35 \\
    x \ge 1
\end{gather*}
