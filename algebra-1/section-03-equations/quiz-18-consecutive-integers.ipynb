> 2. Find two consecutive integers that sum to $45$.

\begin{gather*}
    x + (x + 1) = 45 \\
    2x = 45 - 1 \\
    2x = 44 \\
    \frac{2x = 44}{2} \\ \\

    \boxed{x = 22}
\end{gather*}

The answer is $22, 23$

> 3. In a string of three consecutive integers, the sum of the first two integers is $10$ more than the third integer. What is the third integer?

\begin{gather*}
    x + (x + 1) = (x + 2) + 10 \\
    2x + 1 = x + 12 \\
    2x - x = 12 - 1 \\ \\

    x = 11 \\ \\

    \textbf{CHECKING} \\
    x + (x + 1) = (x + 2) + 10 \\
    (11) + (11 + 1) = (11 + 2) + 10 \\
    11 + 12 = 13 + 10 \\
    23 = 23
\end{gather*}

The answer is $13$
