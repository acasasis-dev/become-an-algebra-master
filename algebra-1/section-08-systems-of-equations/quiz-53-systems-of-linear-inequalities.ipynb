## Quiz 53: Systems of Linear Inequalities

> 1.

$$
    \begin{gather*}
        2x + y > 0 \\
        \boxed{y > -2x + 0}
    \end{gather*}
$$

$$
    \begin{gather*}
        x - y < 3 \\
        -y < -x + 3 \\
        -(-y < -x + 3) \\
        \boxed{y > x - 3}
    \end{gather*}
$$

> **ANSWER: B**

> 2.

$$
    \begin{gather*}
        3x + y \ge 1 \\
        \boxed{y \ge -3x + 1}
    \end{gather*}
$$

$$
    \begin{gather*}
        3x + y \le 4 \\
        \boxed{y \le -3x + 4}
    \end{gather*}
$$

> **ANSWER: A**

> 3.

$$
    \begin{gather*}
        x > y \\
        \boxed{y < x}
    \end{gather*}
$$

$$
    \begin{gather*}
        x + 2y \ge 2 \\
        2y \ge -x + 2 \\
        \frac{2y}{2} \ge \frac{-x + 2}{2} \\
        \boxed{y \ge -\frac{1}{2}x + 1}
    \end{gather*}
$$
