## Quiz 43: Graphing Inequalities On A Number Line

> 1. Graph $x > 1$ on a number line.



The answer is **A** as the line should have an open circle and covers every value that is on the right of point $1$

> 2. Graph $x < - 2$ on a number line.



The answer is **C** as the line should have an open circle and covers every value on the left of point $-2$

> 3. Graph $2x - 1 \ \cancel{>} \ x - 3$ on a number line.



\begin{gather*}
    2x - 1 \ \cancel{>} \ x - 3 \\ \\

    2x - 1 \le x - 3 \\
    2x - x \le -3 + 1 \\ \\

    \boxed{x \le -2}
\end{gather*}

The answer is **A** as the line should have an closed circle and covers every value on the left of point $-2$
