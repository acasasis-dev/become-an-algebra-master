> 2. Solve the inequality.

\begin{gather*}
    4(1 - x) \ \cancel{\le} \ 5(2 - x) \\ \\
    4 - 4x > 10 - 5x \\
    5x - 4x > 10 - 4 \\ \\
    
    \boxed{x > 6}
\end{gather*}

> 3. Solve the inequality.

\begin{gather*}
    -2(3 - x) \ \cancel{\le} \ 3(5 - x) + 4x \\ \\

    -6 + 2x > 15 - 3x + 4x \\
    2x + 3x - 4x > 15 + 6 \\
    5x - 4x > 21 \\ \\

    \boxed{x > 21}
\end{gather*}
