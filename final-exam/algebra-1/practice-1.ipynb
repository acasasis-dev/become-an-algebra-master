> 4. Solve $−(4x − 10) + 14 = 2(4x − 6)$.

\begin{gather*}
    −(4x − 10) + 14 = 2(4x − 6) \\
    -4x + 10 + 14 = 8x - 12 \\
    -4x + 24 = 8x - 12 \\
    8x + 4x = 24 + 12 \\
    12x = 36 \\
    \frac{12x = 36}{12} \\
    \boxed{x = 3} \\ \\

    \textbf{CHECKING} \\
    -(4x - 10) + 14 = 2(4x - 6) \\
    -(4(3) - 10) + 14 = 2(4(3) - 6) \\
    -(12 - 10) + 14 = 2(12 - 6) \\
    -(2) + 14 = 2(6) \\
    12 = 12
\end{gather*}

> 8. Find $(fg)(−3)$ if $f(x) = x − 3$ and $g(x) = 2x + 5$.

\begin{gather*}
    f(x) = x − 3 \\
    f(-3) = (-3) - 3 \\
    f(-3) = -6 \\ \\

    g(x) = 2x + 5 \\
    g(-3) = 2(-3) + 5 \\
    g(-3) = -6 + 5 \\
    g(-3) = -1 \\ \\

    (fg)(-3) = -6 \cdot -1 \\
    \boxed{(fg)(-3) = 6}
\end{gather*}

\begin{gather*}
