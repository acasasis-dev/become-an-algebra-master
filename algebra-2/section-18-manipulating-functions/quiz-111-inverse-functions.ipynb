> 2. Which of these functions is the inverse of the given function?

\begin{gather*}
    f(x) = \frac{1}{x} - 2 \\ \\
\end{gather*}



\begin{gather*}
    y = \frac{1}{x} - 2 \\
    x = \frac{1}{y} - 2 \\
    \frac{1}{y} = x + 2 \\
    y = \frac{1}{x + 2} \\ \\

    \boxed{f^{-1}(x) = \frac{1}{x + 2}}
\end{gather*}

The answer is C

> 3. Which of these is the inverse of the given function?

\begin{gather*}
    f(x) = 3x^3 - 4 \\ \\
\end{gather*}



\begin{gather*}
    y = 3x^3 - 4 \\
    x = 3y^3 - 4 \\
    3y^3 = x + 4 \\
    \frac{3y^3 = x + 4}{3} \\
    y^3 = \frac{x + 4}{3} \\
    \sqrt[3]{y^3 = \frac{x + 4}{3}} \\
    y = \left(\frac{x + 4}{3}\right)^{\frac{1}{3}} \\ \\

    \boxed{f^{-1}(x) = \sqrt[3]{\frac{x + 4}{3}}}
\end{gather*}

The answer is D
