> 2. Which function is not one-to-one?



\begin{gather*}
    j(x) = \sqrt{x + 2} \\
    \sqrt{a + 2} = \sqrt{b + 2} \\
    (\sqrt{a + 2} = \sqrt{b + 2})^2 \\
    a + 2 = b + 2 \\
    a = b + 2 - 2 \\
    a = b \\ \\

    g(x) = 7x - 2 \\
    7a - 2 = 7b - 2 \\
    7a = 7b - 2 + 2 \\
    7a = 7b \\
    a = b \\ \\

    h(x) = (x - 2)(x + 3) \\
    (a - 2)(a + 3) = (b - 2)(b + 3) \\
    a^2 + a - 6 = b^2 + b - 6 \\
    a^2 + a = b^2 + b - 6 + 6 \\
    a^2 + a \ne b^2 + b \\ \\

    f(x) = \frac{x + 7}{x + 5} \\
    \frac{a + 7}{a + 5} = \frac{b + 7}{b + 5} \\
    ab + 5a + 7b + 35 = ab + 5b + 7a + 35 \\
    5a + 7b = 5b + 7a \\
    7b - 5b = 7a - 5a \\
    2b = 2a \\
    a = b
\end{gather*}
