## Quiz 116: Laws of Logarithms

> 1. Write the expression as a rational number if possible, or if not, as a single logarithm.

\begin{gather*}
    \log_3 54 - \log_3 2 \\ \\

    \log_3 \frac{54}{2} \\
    \log_3 27 = x \\
    3^x = 27 \\
    3^x = 3 \cdot 3 \cdot 3 \\
    3^x = 3^3 \\
    \boxed{x = 3}
\end{gather*}

> 2. Which expression is equal to $1$?



\begin{gather*}
    \log_5 20 - \log_5 10 \\
    \log_5 \frac{20}{10} \\
    x = \log_5 2 \\ \\

    \log_3 18 - \log_3 6 \\
    \log_3 \frac{18}{6} \\
    x = \log_3 3 \\
    x = 1 \\ \\

    \log_2 8 - \log_2 7 \\
    x = \log_2 \frac{8}{7} \\ \\

    \log_8 128 - \log_8 2 \\
    \log_8 \frac{128}{2} \\
    x = \log_8 64 \\
    8^x = 64 \\
    8^x = 8^2 \\
    x = 2
\end{gather*}

The answer is B.

> 3. Write the expression as a rational number if possible, or if not, as a single logarithm.

\begin{gather*}
    \log_2 \frac{1}{4} + \log_2 16 \\ \\

    \log_2 \left(\frac{1}{4}(16)\right) \\
    \log_2 \left(\frac{16}{4}\right) \\
    x = \log_2 4 \\
    2^x = 4 \\
    2^x = 2^2 \\ 
    \boxed{x = 2}
\end{gather*}
