## Quiz 95: Systems of Three Equations

> 1. Use any method to find the solution to the system of equations. 

\begin{gather*}
    x + y - z = 4 \\
    x - y - z = 2 \\
    x + 2y + z = 1 \\ \\

    x + y - z = 4 \\
    + (x - y - z = 2) \\ \\

    2x - 2z = 6 \\
    \frac{2x - 2z = 6}{2} \\
    x - z = 3 \\ \\

    2(x - y - z = 2) \\
    + (x + 2y + z = 1) \\ \\

    2x - 2y - 2z = 4 \\
    + (x + 2y + z = 1) \\
    3x - z = 5 \\ \\

    -3(x - z = 3) \\
    + (3x - z = 5) \\ \\

    -3x + 3z = -9 \\
    + (3x - z = 5) \\
    2z = -4 \\
    \frac{2z = -4}{2} \\
    \boxed{z = -2} \\ \\

    x - z = 3 \\
    x - (-2) = 3 \\
    x + 2 = 3 \\
    x = 3 - 2 \\
    \boxed{x = 1} \\ \\

    x + y - z = 4 \\
    (1) + y - (-2) = 4 \\
    1 + y + 2 = 4 \\
    y + 3 = 4 \\
    y = 4 - 3 \\
    \boxed{y = 1} \\ \\

    \textbf{CHECKING} \\
    x + y - z = 4 \\
    (1) + (1) - (-2) = 4 \\
    2 + 2 = 4 \\
    4 = 4 \\ \\

    x - y - z = 2 \\
    (1) - (1) - (-2) = 2\\
    0 + 2 = 2 \\
    2 + 2 \\ \\

    x + 2y + z = 1 \\
    (1) + 2(1) + (-2) = 1 \\
    1 + 2 - 2 = 1 \\
    1 = 1
\end{gather*}

> 3. Solve this system of equations for $x$, $y$, and $z$. 

\begin{gather*}
    x + 2z = 3 \\
    3x - 2y + z = -11 \\
    2x + y + 3z = 9 \\ \\

    x = 3 - 2z \\ \\

    3(3 - 2z) - 2y + z = -11 \\
    9 - 6z - 2y + z = - 11 \\

    6 - 4z + y + 3z = 9 \\
    -z + y = 9 - 6 \\
    -z + y = 3 \\

    -5z - 2y = -20 \\
    + (-z + y = 3) \\ \\

    -5z - 2y = -20 \\
    + 2(-z + y = 3) \\ \\

    -5z - 2y = -20 \\
    + -2z + 2y = 6 \\ \\

    -7z = -14 \\
    \frac{-7z = -14}{-7} \\
    \boxed{z = 2} \\ \\

    x = 3 - 2z \\
    x = 3 - 2(2) \\
    x = 3 - 4 \\
    \boxed{x = -1} \\ \\

    3x - 2y + z = -11 \\
    3(-1) - 2y + (2) = -11 \\
    -3 - 2y + 2 = -11 \\
    -1 - 2y = -11 \\
    -2y = -11 + 1 \\
    -2y = -10 \\
    \frac{-2y = -10}{-2} \\

    x + 2z = 3 \\
    (-1) + 2(2) = 3 \\
