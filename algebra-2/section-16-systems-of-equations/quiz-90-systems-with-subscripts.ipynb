## Quiz 90: Systems with Subscripts

> 1. Use any method to find the unique solution to the system of equations.

\begin{gather*}
    R_{M}T_{M} = 200 \\
    R_{P} = 5R_{M} \\
    R_{P}T_{P} = 500 \\
    T_{P} = 3 - T_{M} \\ \\

    R_{M} = a \\
    T_{M} = b \\
    R_{P} = c \\
    T_{P} = d \\ \\

    ab = 200 \\
    c = 5a \\
    cd = 500 \\
    d = 3 - b \\ \\

    cd = 500 \\
    (5a)(3 - b) = 500 \\
    15a - 5ab = 500 \\
    3a - ab = 100 \\
    ab = 3a - 100 \\ \\

    200 = 3a - 100 \\
    100 + 200 = 3a \\
    300 = 3a \\
    \boxed{a = 100} \\ \\

    (100)b = 200 \\
    \boxed{b = 2} \\ \\

    c = 5a \\
    c = 5(100) \\
    \boxed{c = 500} \\ \\

    d = 3 - b \\
    d = 3 - (2) \\
    \boxed{d = 1} \\ \\

    cd = 500 \\
    (500)(1) = 500 \\
    \boxed{500 = 500}
\end{gather*}

> 2. Solve this system for $R_b$.

\begin{gather*}
    3R_{a} - R_{b} = 15 \\
    2R_{a} + R_{b} = 5 \\ \\

    R_{a} = a \\
    R_{b} = b \\ \\

    3a - b = 15 \\
    2a + b = 5 \\ \\

    3a - b = 15 \\
    + (2a + b = 5) \\ \\

    5a = 20 \\
    \boxed{a = 4} \\ \\

    3(4) - b = 15 \\
    12 - b = 15 \\
    12 - 15 = b \\
    \boxed{b = -3} \\ \\

    \textbf{CHECKING} \\
    3(4) - (-3) = 15 \\
    12 + 3 = 15 \\
    15 = 15 \\ \\

    2(4) + (-3) = 5 \\
    8 - 3 = 5 \\
    5 = 5
\end{gather*}

> 3. Solve this system for $C_p$.

\begin{gather*}
    7C_{p} + 4C_{v} = 2 \\
    -14C_{p} - 12C_{v} = -20 \\ \\

    C_{p} = a \\
    C_{v} = b \\ \\

    7a + 4b = 2 \\
    -14a - 12b = -20 \\ \\

    2(7a + 4b = 2) \\
    + (-14a - 12b = -20) \\ \\

    14a + 8b = 4 \\
    + (-14a - 12b = -20) \\ \\

    -4b = -16 \\
    \boxed{b = 4} \\ \\

    7a + 4(4) = 2 \\
    7a + 16 = 2 \\
    7a = -16 + 2 \\
    7a = -14 \\
    \boxed{a = -2} \\ \\

    \textbf{CHECKING} \\
    7(-2) + 4(4) = 2 \\
    -14 + 16 = 2 \\
    2 = 2 \\ \\

    -14(-2) - 12(4) = -20 \\
    28 - 48 = -20 \\
    -20 = -20
\end{gather*}
