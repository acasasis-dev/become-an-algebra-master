## Quiz 82: Direct Variation

> 1. If **3k = 15** and **kx = 75**, find **x**

$$
    \begin{gather*}
        3k = 15 \\
        \frac{3k = 15}{3} \\
        \frac{\cancel{3}k = \cancel{15}^{\ \ 5}}{\cancel{3}} \\
        \boxed{k = 5} \\ \\
        kx = 75 \\
        (5)x = 75 \\
        5x = 75 \\
        \frac{5x = 75}{5} \\
        \frac{\cancel{5}x = \cancel{75}^{\ \ 15}}{\cancel{5}} \\
        \boxed{x = 15}
    \end{gather*}
$$

> 3. If we attach weights to a hanging spring, it will stretch. The amount of stretch varies directly with the amount of the hanging weight. If a weight of **20** will stretch the spring a distance of **4**, what distance will a weight of **50** stretch it? Let **w** be the weight, **d** be distance, and use the equation **w = kd**

$$
    \begin{gather*}
        w = kd \\
        20 = k(4) \\
        \frac{20 = 4k}{4} \\
        \frac{\cancel{20}^{\ 5} = \cancel{4}k}{\cancel{4}} \\
        \boxed{k = 5} \\ \\
        50 = (5)d \\
        \frac{50 = 5d}{5} \\
        \frac{\cancel{50}^{\ \ 10} = \cancel{5}d}{\cancel{5}} \\
    
        \boxed{d = 10}
    \end{gather*}
$$
