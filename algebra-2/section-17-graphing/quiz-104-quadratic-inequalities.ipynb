## Quiz 104: Quadratic Inequalities

```python
import matplotlib.pyplot as plt
import numpy as np

def plot(*args, ranges=None):
    _, axs = plt.subplots()
    if ranges:
        for i, r in enumerate(ranges):
            start, end = r
            x = np.arange(start, end)
            y = args[i](x)
            plt.plot(x, y)
    else:
        x = np.arange(-20, 20)
        for f in args:
            y = f(x)
            plt.plot(x, y)

    axs.axhline(0, color="black", linewidth=2)
    axs.axvline(0, color="black", linewidth=2)
    plt.xticks(np.arange(-20, 20, 1))
    plt.yticks(np.arange(-20, 20, 1))
    plt.xlim(-20, 20)
    plt.ylim(-20, 20)
    plt.grid()
    return axs, x, y 
```

> 1. Solve $x^2 + 2x - 3 < 0$ graphically.

```python
def f(x):
	return x**2 + 2*x - 3

_plot, x, y = plot(f)
_plot.fill_between(x, 0, y, where=(y < 2), color='red', alpha=0.3)
plt.show()
```

The answer is $\boxed{-3 < x < 1}$

> 2. Solve $x^2 + 10x + 21 > 0$ algebraically.

\begin{gather*}
    x^2 + 10x + 21 = 0 \\
    (x + 7)(x + 3) = 0 \\
    x = -7, -3 \\ \\

    x = -9 \\
    -9^2 + 10(-9) + 21 \\
    81 - 90 + 21 \\
    -9 + 21 \\
    13 \\
    x = -5 \\
    -5^2 + 10(-5) + 21 \\
    25 - 50 + 21 \\
    -25 + 21 \\
    -4 \\
    x = -1 \\
    -1^2 + 10(-1) + 21 \\
    1 - 10 + 21 \\
    -9 + 21 \\
    12 \\ \\

    \boxed{-\infty < x < -7 \ \text{and} -3 < x < \infty}
\end{gather*}

    

> 3. Solve the quadratic inequality $-x^2 + x + 12 \ge 0$.

\begin{gather*}
    -x^2 + x + 12 = 0 \\
    x^2 - x - 12 = 0 \\
    (x - 4)(x + 3) = 0 \\
    x = -3, 4 \\ \\

    x = -5 \\
    -(-5)^2 - 5 + 12 \\
    -(25) - 5 + 12 \\
    -30 + 12 \\
    -18 \\
    x = 0 \\
    -(0)^2 - 0 + 12 \\
    0 - 0 + 12 \\
    0 + 12 \\
    12 \\
    x = 7 \\
    -(7)^2 - 7 + 12 \\
    -49 - 7 + 12 \\
    -56 + 12 \\
    -44 \\ \\

    \boxed{-3 \le x \le 4}
\end{gather*}
