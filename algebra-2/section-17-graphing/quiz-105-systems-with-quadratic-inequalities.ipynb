## Quiz 105: Systems With Quadratic Inequalities

```python
import matplotlib.pyplot as plt
import numpy as np

def plot(*args, ranges=None):
    _, axs = plt.subplots()
    if ranges:
        for i, r in enumerate(ranges):
            start, end = r
            x = np.arange(start, end)
            y = args[i](x)
            plt.plot(x, y)
    else:
        x = np.arange(-20, 20)
        for f in args:
            y = f(x)
            plt.plot(x, y)

    axs.axhline(0, color="black", linewidth=2)
    axs.axvline(0, color="black", linewidth=2)
    plt.xticks(np.arange(-20, 20, 1))
    plt.yticks(np.arange(-20, 20, 1))
    plt.xlim(-20, 20)
    plt.ylim(-20, 20)
    plt.grid()
    return axs, x, y 
```

> 1. Sketch the solution to the system of inequalities.

\begin{gather*}
    y \ge x^2 - 2 \\
    x \le 5y
\end{gather*}



```python
def f_1(x):
    return x**2 - 2

def f_2(x):
    return x / 5

plot(f_1, f_2)
plt.show()
```

The answer is **B**

> 2. Sketch the solution to the system of inequalities.

\begin{gather*}
    x + y < -1 \\
    y \le x^2 + 4x + 3
\end{gather*}



\begin{gather*}
    x + y < - 1 \\
    x + y = -1 \\
    y = -x - 1 \\ \\

    y \le x^2 + 4x + 3 \\
    y = x^2 + 4x + 3 \\ \\
\end{gather*}
    

```python
def f_1(x):
	return -x - 1

def f_2(x):
	return x**2 + 4*x + 3

plot(f_1, f_2)
plt.show()
```

The answer is **B**

> 3. Sketch the solution to the system of inequalities.

\begin{gather*}
    x - 2y - 2 \le 0 \\
    x^2 + 2y > 4 
\end{gather*}



\begin{gather*}
    x - 2y - 2 \le 0 \\
    x - 2y - 2 = 0 \\
    2y = x - 2 \\
    \frac{2y = x - 2}{2} \\
    y = \frac{x - 2}{2} \\ \\

    x^2 + 2y > 4 \\
    x^2 + 2y = 4 \\
    2y = -x^2 + 4 \\
    \frac{2y = -x^2 + 4}{2} \\
    y = \frac{-x^2 + 4}{2}
\end{gather*}

```python
def f_1(x):
	return (x - 2) / 2

def f_2(x):
	return (-x**2 + 4) / 2

plot(f_1, f_2)
plt.plot()
```
