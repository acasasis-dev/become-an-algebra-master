## Quiz 102: Modeling A Piecewise-Defined Function

> 1. What is the definition of the piecewise function shown in the graph?



\begin{gather*}
    (-1, 4) \ \& \ (1, -4) \\
    m = \frac{y_2 - y_1}{x_2 - x_1} \\
    m = \frac{-4 - 4}{1 - (-1)} \\
    m = \frac{-8}{2} \\
    m = -4 \\ \\

    y - y_1 = m(x - x_1) \\
    y - 4 = -4(x - (-1)) \\
    y - 4 = -4(x + 1) \\
    y - 4 = -4x - 4 \\
    y = -4x - 4 + 4 \\
    y = -4x \\ \\


    \boxed{f(x) = \begin{cases}
        4 &\text{if } x \le -1 \\
        -4x &\text{if } -1 \le x \le 1 \\
        -4 &\text{if } x \ge 1
    \end{cases}}
\end{gather*}

> 2. What is the definition of the piecewise function shown in the graph?



\begin{gather*}
    (-4, -3) \ \& \ (3, 4) \\
    m = \frac{y_2 - y_1}{x_2 - x_1} \\
    m = \frac{4 - (-3)}{3 - (-4)} \\
    m = \frac{4 + 3}{3 + 4} \\
    m = \frac{7}{7} \\
    m = 1  \\ \\

    y - y_1 = m(x - x_1) \\
    y - (-3) = 1(x - (-4)) \\
    y + 3 = x + 4 \\
    y = x + 4 - 1 \\
    y = x + 1 \\ \\

    (3, 4) \ \& \ (7, 0) \\
    m = \frac{0 - 4}{7 - 3} \\
    m = \frac{-4}{4} \\
    m = -1 \\ \\

    y - y_1 = m(x - x_1) \\
    y - 4 = -1(x - 3) \\
    y - 4 = -x + 3 \\
    y = -x + 3 + 4 \\
    y = -x + 7 \\ \\

    \boxed{f(x) = \begin{cases}
        -3 &\text{if } x < -4\\
        x + 1 &\text{if } -4 < x < 3\\
        -x + 7 &\text{if } x > 3\\
    \end{cases}}
\end{gather*}
