## Quiz 97: Graphing Parabolas

> 1. Write the equation in vertex form.

\begin{gather*}
    y = x^2 + 8x + 5 \\ \\

    (\frac{8}{2})^2 = (4)^2 = 16 \\ \\

    y + 16 = (x^2 + 8x + 16) + 5 \\
    y + 16 = (x + 4)^2 + 5 \\
    y = (x + 4)^2 + 5 - 16 \\
    y = (x + 4)^2 - 11 
\end{gather*}

> 3. Which parabola is the graph of $y = x^2 - 4x + 2$?



\begin{gather*}
    y = x^2 - 4x + 2 \\
    y = (0)^2 - 4(0) + 2 \\
    y = 2 \\ \\

    (\frac{4}{2})^2 = (2)^2 = 4 \\ \\

    y = x^2 - 4x + 2 \\
    y = (x^2 - 4x + 4) - 4 + 2 \\
    y = (x - 2)^2 - 2 \\
\end{gather*}

**ANSWER: D**
