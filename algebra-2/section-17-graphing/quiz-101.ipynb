## Quiz 101: Equation Modeling

> 1. A car and a truck were driven for a week. The car traveled $400$ miles more than the truck. The fuel mileage for the car was $27$ miles per gallon (mpg), and the fuel mileage for the truck was $18$ mpg. Write an equation which gives the total amount of fuel, $g$(in gallons), that was used by the two vehicles that week in terms of $c$, the distance (in miles) traveled by the car.



> 2. The average speed of any falling object between time $0$ (the time at which it starts falling) and time $t$ is given by the ratio of the distance $d$ through which it has fallen to the elapsed time $(t)$, $V = D|t$. A ball is thrown at a speed of $12 \frac{f}{s}$ straight downward from a tall cliff. The distance through which it has fallen is $D = 16t^2 + 12t$, where $t$ is the time (in seconds) that it's been falling. Write an equation that gives $t$ in terms of $V$.

> 3. At a movie theater, prices are $\$4$ for children and $\$8$ for adults. If $800$ people came to see a movie, write an equation that gives the number of children, $c$, in terms of $T$, the total amount of money taken in.
