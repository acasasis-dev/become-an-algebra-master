> 2. What is the equation of the given circle?



Center is at $(h, k) = (0, 5)$ and radius is $3$ so the equation is 

\begin{gather*}
    (x - 0)^2 + (y - 5)^2 = 9 \\
    x^2 + (y - 5)(y - 5) = 9 \\
    x^2 + y^2 - 5y - 5y + 25 = 9 \\
    x^2 + y^2 - 10y + 25 + 9 = 0 \\
    \boxed{x^2 + y^2 - 10y + 16 = 0}
\end{gather*}

> 3. What is the equation of the given circle?



Center is at $(h, k) = (-3, 2)$ and radius is $5$ so the equations is 

\begin{gather*}
    (x + 3)^2 + (y - 2)^2 = 25 \\
    (x + 3)(x + 3) + (y - 2)(y - 2) = 25 \\
    (x^2 + 3x + 3x + 9) + (y^2 - 2y - 2y + 4) = 25 \\
    x^2 + 6x + 9 + y^2 - 4y + 4 = 25 \\
    x^2 + 6x + y^2 - 4y = 25 - 9 - 4 \\
    \boxed{x^2 + 6x + y^2 - 4y = 12}
\end{gather*}
